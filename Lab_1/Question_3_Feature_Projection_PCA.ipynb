```python
import numpy as np
```

```python
X = np.loadtxt("F:\\Downloads\\Compressed\\Animals_with_Attributes2\\Features\\ResNet101\\AwA2-features.txt")
y = np.loadtxt("F:\\Downloads\\Compressed\\Animals_with_Attributes2\\Features\\ResNet101\\AwA2-labels.txt")
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 0, stratify = y)
```

# Preprocessing: normalize using MinMaxScaler

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(.95).fit(X_train)
```

```python
X_train_new = pca.transform(X_train)
```

```python
X_train_new.shape
```

```python
clf = SVC(kernel = 'linear', C = 0.001).fit(X_train_new, y_train) 
```

```python
X_test_new = pca.transform(X_test)
```

```python
X_test_new.shape
```

```python
from sklearn import metrics
y_pred = clf.predict(X_test_new)
y_true = y_test
```

```python
accuracy = metrics.accuracy_score(y_true, y_pred)
print(accuracy)
```

```python
np.set_printoptions(threshold=np.inf)
metrics.confusion_matrix(y_true, y_pred)
```
